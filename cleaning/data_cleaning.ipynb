```python
import json
```

```python
# Load raw data from file
with open("raw_data.json", "r") as f:
    raw_data = json.load(f)

print("raw_data loaded successfully")
```

```python
def clean_data(raw_data):
    # 1. Remove users with empty names
    raw_data['users'] = [
        user for user in raw_data['users']
        if user['name'].strip()
    ]

    # 2. Remove duplicate friends
    for user in raw_data['users']:
        user['friends'] = list(set(user['friends']))

    # 3. Keep users who have friends OR liked pages
    raw_data['users'] = [
        user for user in raw_data['users']
        if user['friends'] or user['liked_pages']
    ]

    # 4. Remove duplicate pages (by page id)
    unique_pages = {}
    for page in raw_data['pages']:
        unique_pages[page['id']] = page

    raw_data['pages'] = list(unique_pages.values())

    return raw_data

# Clean the Data
data = clean_data(raw_data)

with open("cleaned_data.json", "w") as f:
    json.dump(data, f, indent=4)

print("cleaned_data.json created successfully")
```
