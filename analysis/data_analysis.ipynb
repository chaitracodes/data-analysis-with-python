```python
import json
```

```python
# Function to load the data
def load_data(filename):
  with open(filename,"r") as f:
    data = json.load(f)
  return data
```

```python
# check if the data loads
data = load_data("..data/cleaned/data.json")
```

```python
# to display users and their connections
def display_users(data):

  print("Users and their Connections:")
  for user in data['users']:
    print(f"ID:{user['id']} - {user['name']} is friends with {user['friends']} and the liked pages are {user['liked_pages']}")

  print("\nPages Info:")
  for page in data['pages']:
    print(f"{page['id']} {page['name']}")

# to load and display
data = load_data("data.json")
display_users(data)
```
