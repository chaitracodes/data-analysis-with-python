```python
import json

# Load large dataset
def load_data(filename):
    with open(filename, "r") as f:
        return json.load(f)

# People You May Know logic
def find_people_you_may_know(user_id, large_raw_data):
    # Build user -> friends mapping
    user_friends = {}
    for user in large_raw_data['users']:
        user_friends[user['id']] = set(user['friends'])

    # If user not found
    if user_id not in user_friends:
        return []

    # Direct friends of the given user
    direct_friends = user_friends[user_id]
    suggestions = {}

    # Find mutual friends
    for friend in direct_friends:
        for mutual in user_friends.get(friend, []):
            # Exclude the user themself and existing friends
            if mutual != user_id and mutual not in direct_friends:
                suggestions[mutual] = suggestions.get(mutual, 0) + 1

    # Sort by mutual friend count (descending)
    sorted_suggestions = sorted(
        suggestions.items(),
        key=lambda x: x[1],
        reverse=True
    )

    # Return only user IDs
    return [uid for uid, _ in sorted_suggestions]


# Load dataset (GitHub-correct path)
large_raw_data = load_data("large_raw_data.json")

# Example usage
user_id = 1
recommendations = find_people_you_may_know(user_id, large_raw_data)

print(f"People You May Know for User {user_id}: {recommendations}")
```
